```python

#  ██╗███╗   ███╗██████╗  ██████╗ ██████╗ ████████╗██╗███╗   ██╗ ██████╗ 
#  ██║████╗ ████║██╔══██╗██╔═══██╗██╔══██╗╚══██╔══╝██║████╗  ██║██╔════╝ 
#  ██║██╔████╔██║██████╔╝██║   ██║██████╔╝   ██║   ██║██╔██╗ ██║██║  ███╗
#  ██║██║╚██╔╝██║██╔═══╝ ██║   ██║██╔══██╗   ██║   ██║██║╚██╗██║██║   ██║
#  ██║██║ ╚═╝ ██║██║     ╚██████╔╝██║  ██║   ██║   ██║██║ ╚████║╚██████╔╝
#  ╚═╝╚═╝     ╚═╝╚═╝      ╚═════╝ ╚═╝  ╚═╝   ╚═╝   ╚═╝╚═╝  ╚═══╝ ╚═════╝ 
#        

id_dict = {}
id_dict[1] = []
id_dict[2] = []
id_dict[3] = []



#importing and cleaning data

#load data function adapted from practical session
def load_data(i_file,no_class=False):
    
    #defining X and Y as emptys
    X = []
    Y = []
    skip_first = True
    
    #open and starts reading data
    with open(i_file, mode='r') as fin:
        for line in fin:
            
            #skips the first line as it would be col names and is not useful
            if(skip_first):
                skip_first=False
                continue
                
            #splits up the CSV line
            atts = line.strip().split(",")
            
            #converts date stamp into unix time int then makes it a does a crude binning method to group like dates
            atts[0] = round(time.mktime(datetime.datetime.strptime(atts[0], "%m/%d/%Y").timetuple())/(30*24*60*60))
            
            #converts ratings into ints
            for i in [4,5,6]:
                atts[i] = int(atts[i])
                
            for i in [1,2,3]:
                if atts[i] in id_dict[i]:
                    atts[i] = id_dict[i].index(atts[i])
                else:
                    temp = atts[i]
                    atts[i] = len(id_dict[i])
                    id_dict[i].append(temp)
                    
            
            #if theres no class, skips appending the class as it doesnt exist
            if(no_class):
                X.append(atts[:]) #all attributess 
            else:
                X.append(atts[:-1]) #all attributes minus the class
                Y.append(int(atts[-1])) #just the class 
    
    #returning depending on if there is a class col or not
    if(no_class):
        return X
    else:
        return X, Y

#adapted load function for loading and cleaning reviews
def load_text(i_file):
    
    #sets empty array
    X = []
    skip_first = True
    
    #open and starts reading data
    with open(i_file, mode='r') as fin:
        for line in fin:
            
            #removes the \n char and the open quote mark
            line = line[1:-1] 
            
            #if the line is nothing, skip it
            if(skip_first or line ==''):
                skip_first=False
                continue
                
            #append whats left
            X.append(line) #all atts 
            
    #returning array
    return X


#Naiming scheme for each data set follows
#
#data set names follow the following convention:
    #trn_***_full = original label data
    #trn_***      = split labelled data for trianing
    #tst1_***     = split labelled data for testing (validation set)
    #tst2_***     = full unlabbeled testing data set 
#
#where *** represent what part of the data set the array is, ** can be
    #att          = attribute array
    #rvw          = review string array
    #cls          = class array
#

#importing csv files
#supervised
trn_att_full, trn_cls_full = load_data('data/review_meta_train.csv')
trn_rvw_full = load_text('data/review_text_train.csv')
#unsupervised data
tst2_att = load_data('data/review_meta_test.csv',no_class=True)
tst2_rvw = load_text('data/review_text_test.csv')


#splitting supervised set
#if splitting into training and validation data sets
splitting = True
if(splitting):
    trn_att, tst1_att, trn_cls, tst1_cls, trn_rvw, tst1_rvw,  = train_test_split(
        trn_att_full, 
        trn_cls_full,
        trn_rvw_full,
        train_size=0.80, 
        random_state=1)
else:
    #If testing on training set is needed, this will set the training and testing data sets to be the same
    tst1_att = trn_att_full
    tst1_rvw = trn_rvw_full
    tst1_cls = trn_cls_full
    trn_att = trn_att_full
    trn_rvw = trn_rvw_full
    trn_cls = trn_cls_full

```
